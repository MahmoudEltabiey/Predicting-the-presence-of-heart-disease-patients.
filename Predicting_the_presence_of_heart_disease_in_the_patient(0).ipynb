Preprocessing the data by handling missing values and encoding categorical features.

```python
#Imputing before encoding to improve accuracy
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')  # median is better than mean for skewed cols
X[['ca', 'thal']] = imputer.fit_transform(X[['ca', 'thal']])

df_encoded = pd.get_dummies(X, columns=['thal', 'cp']) # Encoding categorical features using one hot encoding

print("\nEncoded Dataframe:\n", df_encoded.head())
df.fillna(df.mean(), inplace=True) # handling missing values with the mean of each column

X_encoded = pd.get_dummies(X, columns=['thal', 'cp', 'restecg', 'slope'])


# Checking for missing values again
missing_values = df.isnull().sum()
print("\nMissing Values After Imputation:\n", missing_values)
```

Split the preprocessed data into training and testing sets (80% training, 20% testing)

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.model_selection import train_test_split
# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y.values.ravel(), test_size=0.2, random_state=42, stratify=y
)
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

Implementing Random Forest Classifier algorithm to build a heart disease prediction model and training the model on the training data

```python
#Running the model

model = RandomForestClassifier(max_depth=5)

model.fit(X_train, y_train)
# Flattening y_train to a 1D array to avoid DataConversionWarning
y_train_1d = y_train.ravel()

model.fit(X_train, y_train_1d)
y_predict = model.predict(X_test)
y_pred_quant = model.predict_proba(X_test)[:, 1]
y_pred = model.predict(X_test)
```

Checking Accuracy:

```python
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation on the training set using flattened y
scores = cross_val_score(model, X_train, y_train.ravel(), cv=10)

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Average cross-validation score:", scores.mean())
```

Evaluating the model's performance using appropriate classification / regression metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

# Calculate the mean absolute error
MAbE = mean_absolute_error(y_test, y_pred)
print("Mean absolute error:", MAbE)
```

```python
# Calculate the mean squared error
MSqE = mean_squared_error(y_test, y_pred)
print("Mean squared error:", MSqE)
```

```python
# Calculate the confusion matrix
ConfMat = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", ConfMat)
```

```python
# Calculate the accuracy
Accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", Accuracy)
```

```python
# Calculate the recall
Recall = recall_score(y_test, y_pred, average='weighted')
print("Recall:", Recall)
```

```python
# Calculate the precision
Precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", Precision)
```

```python
# Calculate the f-measure
FMeasure = f1_score(y_test, y_pred, average='weighted')
print("F-measure:", FMeasure)
```

Fine-tuning the model by adjusting hyperparameters and trying different classification / regression algorithms.

```python
# Fine-tuning the model using Random forrest
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(n_estimators=200, max_depth=None, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Calculate the accuracy of the best model
accuracy_best = accuracy_score(y_test, y_pred)

# Print the accuracy of the best model
print("Best Model Accuracy:", accuracy_best)
```
